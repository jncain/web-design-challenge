```python
# import dependencies
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file = "NAEPData.csv"
```

```python
naep = pd.read_csv(file)
naep
```

```python
naep.dtypes
```

```python
math_four = naep.loc[(naep["Subject"]=="Mathematics") & (naep["Grade"] == 4)]
math_eight = naep.loc[(naep["Subject"]=="Mathematics") & (naep["Grade"] == 8)]
read_four = naep.loc[(naep["Subject"]=="Reading") & (naep["Grade"] == 4)]
read_eight = naep.loc[(naep["Subject"]=="Reading") & (naep["Grade"] == 8)]
```

```python
# Graph Math 4th Grade
m4_np = math_four.loc[(math_four["Jurisdiction"]=="National public")]
m4_mn = math_four.loc[(math_four["Jurisdiction"]=="Minnesota")]

xlabels = m4_mn['Year']
npcolor = '#c69214'
mncolor = '#001871'
fig, ax = plt.subplots(figsize=(14, 8))

mn, = plt.plot(m4_mn['Year'], m4_mn["atabPro"], marker='o', markerfacecolor=mncolor, markeredgecolor=mncolor, markersize=8, color=mncolor, linewidth=4, label="Minnesota")
np, = plt.plot(m4_np['Year'], m4_np["atabPro"], marker='o', markerfacecolor=npcolor, markeredgecolor=npcolor, markersize=8, color=npcolor, linewidth=4, label="National public")

plt.tick_params(axis='both', labelsize=14)
plt.ylim(0, 100)
plt.xticks(xlabels)

leg = plt.legend(fontsize=16)
leg.set_title('Jurisdiction:', prop={'size':16, 'weight':'bold'})
plt.title("Results for 4th Grade Students on NAEP Mathematics", fontsize=20, fontweight='bold')
plt.xlabel("Year", fontsize=18, fontweight='bold')
plt.ylabel("Percent of Students\n at or above NAEP Proficient", fontsize=18, fontweight='bold')

plt.savefig("WebVisualizations/assets/Grade4Math.png")

plt.show()
```
